```python
import numpy as np 
import matplotlib.pyplot as plt
# for labeling data in graph
import pandas as pd 
# used to import files
from sklearn.linear_model import LinearRegression
#to import formulas
from sklearn.model_selection import train_test_split 
```

```python
file = pd.read_csv('Mall_customers.csv')
file.head()
```

```python
x = file[['Age']]
y = file['Spending Score (1-100)']
plt.scatter(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
linearrig=LinearRegression()
linearrig.fit(x_train,y_train)
predictions = linearrig.predict(x_test)
print(predictions)
```

```python
print(linearrig.coef_)
```

```python
print(linearrig.intercept_)
```

```python
for actual,pre in zip(y_test,predictions):
    print(f"Actual:{actual},Prediction:{pre}")
    
```

```python
plt.scatter(y_test,predictions,color="green")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.title("Actual vs Prediction")
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Plot actual values in red
plt.scatter(range(len(y_test)), y_test, color='red', label="Actual Values", marker='o')

# Plot predicted values in blue
plt.scatter(range(len(predictions)), predictions, color='blue', label="Predicted Values", marker='x')

# Labels and Title
plt.xlabel("Data Point Index")
plt.ylabel("Annual Income (k$)")
plt.title("Actual vs. Predicted Values in Multiple Linear Regression")

# Show legend
plt.legend()

# Show plot
plt.show()
```
