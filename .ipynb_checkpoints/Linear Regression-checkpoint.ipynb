```python
import numpy as np 
import matplotlib.pyplot as plt
# for labeling data in graph
import pandas as pd 
# used to import files
from sklearn.linear_model import LinearRegression
#to import formulas
from sklearn.model_selection import train_test_split  
# Import train_test_split
# Linear Regression :- used to find mean of Actual and Predicted Value
```

```python
#Importing File by using pandasS 
file = pd.read_csv('Mall_customers.csv')
file.head()
```

```python
x = file[['Age']]  
y = file['Annual Income (k$)']
plt.scatter(x,y,color ="blue")
plt.xlabel("Age")
plt.ylabel("Annual income")
plt.title("Age and Anual income")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
LinearReg = LinearRegression()
LinearReg.fit(x_train, y_train)
prediction = LinearReg.predict(x_test)
print(prediction)
#Predicted values
```

```python
plt.scatter(x_test, y_test, color='red', label="Actual Data")
plt.plot(x_test, prediction, color='green', label="Predicted Data", marker="+")
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.title("Actual vs. Predicted Annual Income")
plt.show()
```
